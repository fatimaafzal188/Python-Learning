### HOW TO DEAL WITH MISSING VALUES IN EDA
### CATEGORICAL/NUMERICAL VARIABLES/OBJECTS/FLOAT/INT/BOOLEAN


```python
df['age'].unique()
```

```python
df['age'].min()
```

```python
df['age'].max()
```

firstly, we are going to creat another variable that consist of and to fill missing numbers in age, we are going to take a mean of it .

```python
df_2=df['age'].fillna(df['age'].mean())
df_2
```

the code written below shows that there is no missing value left. and the above stats showa that all the missing values are replaced by its mean

```python

df_2.isnull().sum()
```

```python
print(df['age'].min())
print(df['age'].max())
print(df['age'].mean())     # it will show mean value of age column (value which is used to fill the missing values)
```

```python
df_2=df
```

```python
df_2['age'].fillna(df_2['age'].mean(),inplace=True)
```

```python
df_2
```

```python
df_3=df_2
```

```python
df_3.isnull().sum()
```

# removing deck because it has many null values

```python
df_3.drop(columns=['deck'], errors='ignore').isnull().sum()
```

```python
df_4=df_3
```

```python
df_4['embark_town'] = df_4['embark_town'].fillna(df_4['embark_town'].mode()[0])
df_4['embarked'] = df_4['embarked'].fillna(df_4['embarked'].mode()[0])
```
