```python
import pandas as pd
import numpy as np
import seaborn as sns

# Universe U and beta
U = ['μ1', 'μ2', 'μ3', 'μ4', 'μ5', 'μ6', 'μ7', 'μ8']
beta = 0.50

# Fuzzy β-coverings C1 and C2
C1 = {
    'C1': {'μ1': 0.5, 'μ2': 0.4133, 'μ3': 0.123, 'μ4': 0.5399, 'μ5': 0.1588, 'μ6':0.119, 'μ7': 0.3857, 'μ8': 0.1206},
    'C2': {'μ1': 0.5867, 'μ2': 0.5, 'μ3': 0.2097, 'μ4': 0.6266, 'μ5': 0.2455, 'μ6': 0.2057, 'μ7': 0.4724, 'μ8': 0.2073},
    'C3': {'μ1':0.877, 'μ2': 0.7903, 'μ3': 0.5, 'μ4':0.9169, 'μ5': 0.5358, 'μ6': 0.496, 'μ7': 0.7627, 'μ8': 0.4976},
    'C4': {'μ1': 0.4601, 'μ2': 0.3734, 'μ3': 0.0831, 'μ4': 0.5, 'μ5': 0.1189, 'μ6': 0.0791, 'μ7': 0.3458, 'μ8': 0.0807},
    'C5': {'μ1': 0.8412, 'μ2': 0.7545, 'μ3': 0.4642, 'μ4': 0.8811, 'μ5': 0.5, 'μ6': 0.4602, 'μ7': 0.7268, 'μ8': 0.503},
    'C6': {'μ1': 0.881, 'μ2': 0.7943, 'μ3': 0.504, 'μ4': 0.9209, 'μ5': 0.5398, 'μ6': 0.5, 'μ7': 0.7667, 'μ8':0.4967},
    'C7': {'μ1': 0.6143, 'μ2':0.5276, 'μ3': 0.2373, 'μ4': 0.6542, 'μ5': 0.2732, 'μ6': 0.2333, 'μ7': 0.5, 'μ8': 0.2349},
    'C8': {'μ1': 0.8794, 'μ2': 0.7927, 'μ3': 0.5024, 'μ4': 0.9193, 'μ5': 0.497, 'μ6': 0.5033, 'μ7': 0.7651, 'μ8': 0.5}
}
C2 = {
    'C1': {'μ1': 0.5, 'μ2':0.3078, 'μ3': 0.2795, 'μ4': 0.5789, 'μ5': 0.4613, 'μ6': 0.2708, 'μ7': 0.7545, 'μ8': 0.2729},
    'C2': {'μ1': 0.6922, 'μ2': 0.5, 'μ3': 0.4716, 'μ4': 0.7711, 'μ5': 0.6535, 'μ6': 0.463, 'μ7': 0.9467, 'μ6': 0.4651},
    'C3': {'μ1': 0.7205, 'μ2': 0.5284, 'μ3': 0.5, 'μ4':0.7995, 'μ5': 0.6819, 'μ6':0.5104, 'μ7': 0.975, 'μ8': 0.4935},
    'C4': {'μ1': 0.4211, 'μ2': 0.2289, 'μ3': 0.2005, 'μ4': 0.5, 'μ5': 0.3824, 'μ6': 0.1919, 'μ7': 0.6756, 'μ8': 0.194},
    'C5': {'μ1': 0.5387, 'μ2': 0.3465, 'μ3': 0.3181, 'μ4': 0.6176, 'μ5':0.5, 'μ6': 0.3095, 'μ7': 0.7932, 'μ8': 0.3116},
    'C6': {'μ1': 0.7292, 'μ2':0.537, 'μ3': 0.4896, 'μ4': 0.8081, 'μ5': 0.6905, 'μ6': 0.5, 'μ7':0.9837, 'μ8': 0.5021},
    'C7': {'μ1': 0.2455, 'μ2': 0.0533, 'μ3': 0.025, 'μ4': 0.3244, 'μ5': 0.2068, 'μ6': 0.0163,  'μ7': 0.5, 'μ8': 0.0184},
    'C8': {'μ1':0.7271, 'μ2': 0.5349, 'μ3': 0.5065, 'μ4': 0.806, 'μ5': 0.6884, 'μ6': 0.4979, 'μ7': 0.9816, 'μ8': 0.5}
}

C3 = {
    'C1': {'μ1': 0.5,    'μ2': 0.3513, 'μ3': 0.3396,  'μ4': 0.4285, 'μ5': 0.3097, 'μ6': 0.2613,  'μ7': 0.5576, 'μ8': 0.3617},
    'C2': {'μ1': 0.6487, 'μ2': 0.5,    'μ3': 0.4883, 'μ4': 0.5772, 'μ5': 0.4584, 'μ6': 0.41, 'μ7': 0.7063, 'μ8': 0.5104},
    'C3': {'μ1': 0.6604,  'μ2': 0.5117, 'μ3': 0.5,    'μ4': 0.5889, 'μ5': 0.4701, 'μ6': 0.4217,  'μ7': 0.7179, 'μ8': 0.5221},
    'C4': {'μ1': 0.5715, 'μ2': 0.4228, 'μ3': 0.4111, 'μ4': 0.5,    'μ5': 0.3812, 'μ6': 0.3329, 'μ7': 0.6291, 'μ8': 0.4332},
    'C5': {'μ1': 0.6903, 'μ2': 0.5416, 'μ3': 0.5299, 'μ4': 0.6188, 'μ5': 0.5,    'μ6': 0.4516, 'μ7': 0.7478, 'μ8': 0.552},
    'C6': {'μ1': 0.7387,  'μ2': 0.59, 'μ3': 0.5783,  'μ4': 0.6671, 'μ5': 0.5484, 'μ6': 0.5,    'μ7': 0.7962, 'μ8': 0.6004},
    'C7': {'μ1': 0.4424, 'μ2': 0.2937, 'μ3': 0.2821, 'μ4': 0.3709, 'μ5': 0.2522, 'μ6': 0.2038, 'μ7': 0.5,    'μ8': 0.3042},
    'C8' : {'μ1': 0.6383, 'μ2': 0.4896, 'μ3': 0.4779, 'μ4': 0.5668, 'μ5': 0.448,  'μ6': 0.3996, 'μ7': 0.6958, 'μ8': 0.5}
}
C4 = { 
    'C1': {'μ1': 0.5,    'μ2': 0.3489, 'μ3': 0.1419,  'μ4': 0.4628, 'μ5': 0.3728, 'μ6': 0.1381,  'μ7': 0.3657, 'μ8': 0.147},
    'C2': {'μ1': 0.6511, 'μ2': 0.5,    'μ3': 0.293, 'μ4': 0.6138, 'μ5': 0.5239, 'μ6': 0.2891, 'μ7': 0.5167, 'μ8': 0.298},
    'C3': {'μ1': 0.8581,   'μ2': 0.707, 'μ3': 0.5,    'μ4': 0.8208, 'μ5': 0.7308, 'μ6': 0.4961,   'μ7': 0.7237, 'μ8': 0.5221},
    'C4': {'μ1': 0.5372, 'μ2': 0.3862, 'μ3': 0.1792, 'μ4': 0.5,    'μ5': 0.41,   'μ6': 0.1753, 'μ7': 0.4029, 'μ8': 0.1842},
    'C5': {'μ1': 0.6272, 'μ2': 0.4761, 'μ3': 0.2692, 'μ4': 0.59,   'μ5': 0.5,   'μ6': 0.2653, 'μ7': 0.4929, 'μ8': 0.2742},
    'C6': {"μ1": 0.8619,  "μ2":0.7109, "μ3": 0.5039,  "μ4": 0.8247, "μ5": 0.7347, "μ6": 0.5,    "μ7": 0.7276, "μ8": 0.5089},
    'C7': {'μ1': 0.6343,   'μ2': 0.4833, 'μ3': 0.2763,   'μ4': 0.5971,   'μ5': 0.5,   'μ6': 0.2724,   'μ7': 0.5,   'μ8': 0.28},
    'C8': {'μ1': 0.853, 'μ2': 0.702, 'μ3': 0.495, 'μ4': 0.8158, 'μ5': 0.7258,  'μ6': 0.4911, 'μ7': 0.7187, 'μ8': 0.5}
}

C5 = {
    'C1': {'μ1': 0.5,    'μ2': 0.3206, 'μ3': 0.2529,  'μ4': 0.3525, 'μ5': 0.3942, 'μ6': 0.3567,  'μ7': 0.3188, 'μ8': 0.2963},
    'C2': {'μ1': 0.6794, 'μ2': 0.5,    'μ3': 0.4323, 'μ4': 0.5319, 'μ5': 0.5736, 'μ6': 0.5361, 'μ7': 0.4982, 'μ8': 0.4758},
    'C3': {'μ1': 0.7471,  'μ2': 0.5677, 'μ3': 0.5,    'μ4': 0.5996, 'μ5': 0.6413, 'μ6': 0.6038,  'μ7': 0.5659, 'μ8': 0.5434},
    'C4': {'μ1': 0.6475, 'μ2': 0.4681, 'μ3': 0.4004, 'μ4': 0.5,    'μ5': 0.5417, 'μ6': 0.5042, 'μ7': 0.4663, 'μ8': 0.4438},
    'C5': {'μ1': 0.6058, 'μ2': 0.4264, 'μ3': 0.3587, 'μ4': 0.4583, 'μ5': 0.5,    'μ6': 0.4625, 'μ7': 0.4246, 'μ8': 0.4021},
    'C6': {'μ1': 0.6433,  'μ2':0.4639, 'μ3': 0.3962,  'μ4': 0.4958, 'μ5': 0.5375, 'μ6': 0.5,    'μ7': 0.4621, 'μ8': 0.4396},
    'C7': {'μ1': 0.6812, 'μ2': 0.5018, 'μ3': 0.4341, 'μ4': 0.5337, 'μ5': 0.5754, 'μ6': 0.5379, 'μ7': 0.5,    'μ8': 0.4775},
    'C8': {'μ1': 0.7037, 'μ2': 0.5242, 'μ3': 0.4566, 'μ4': 0.5562, 'μ5': 0.5979,  'μ6': 0.5604, 'μ7': 0.5225, 'μ8': 0.5}
}
C6 = {
    'C1': {'μ1': 0.5,    'μ2': 0.4657, 'μ3': 0.1209,  'μ4': 0.4407, 'μ5': 0.3061, 'μ6': 0.1183, 'μ7': 0.436,'μ8': 0.1252},
    'C2': {'μ1': 0.5343, 'μ2': 0.5,    'μ3': 0.1552, 'μ4': 0.4749, 'μ5': 0.3404, 'μ6': 0.1526, 'μ7': 0.4703, 'μ8': 0.1594},
    'C3': {'μ1': 0.8791,  'μ2': 0.8448, 'μ3': 0.5,    'μ4': 0.8198, 'μ5': 0.6852, 'μ6': 0.4974,  'μ7': 0.8151, 'μ8': 0.5043},
    'C4': {'μ1': 0.5593, 'μ2': 0.5251, 'μ3': 0.1802, 'μ4': 0.5,    'μ5': 0.3654, 'μ6': 0.1777, 'μ7': 0.4953, 'μ8': 0.1845},
    'C5': {'μ1': 0.6939, 'μ2': 0.6596, 'μ3': 0.3148, 'μ4': 0.6346, 'μ5': 0.5,    'μ6': 0.3122, 'μ7': 0.6299, 'μ8': 0.3191},
    'C6': {"μ1": 0.8817,  "μ2":0.8474, "μ3": 0.5026,  "μ4": 0.8223, "μ5": 0.6878, "μ6": 0.5,    "μ7": 0.8176, "μ8": 0.5068},
    'C7': {'μ1': 0.564, 'μ2': 0.5297, 'μ3': 0.1849, 'μ4': 0.5047, 'μ5': 0.3701, 'μ6': 0.1824, 'μ7': 0.5,    'μ8': 0.1892},
    'C8': {'μ1': 0.8748, 'μ2': 0.8406, 'μ3': 0.4957, 'μ4': 0.8155, 'μ5': 0.6809,  'μ6': 0.4932, 'μ7': 0.8108, 'μ8': 0.5}
}

C7= {
    'C1': {'μ1': 0.5,   'μ2': 0.2156, 'μ3': 0.1706,  'μ4': 0.4826, 'μ5': 0.5192, 'μ6': 0.167,  'μ7': 0.5129, 'μ8': 0.1726},
    'C2' : {'μ1': 0.7844, 'μ2': 0.5,    'μ3':0.455, 'μ4': 0.767, 'μ5': 0.8036, 'μ6': 0.4514, 'μ7': 0.7973, 'μ8': 0.457},
    'C3' : {'μ1': 0.8294,  'μ2': 0.545, 'μ3': 0.5,    'μ4': 0.812, 'μ5':0.8486, 'μ6': 0.4964,  'μ7': 0.8423, 'μ8': 0.502},
    'C4' : {'μ1': 0.5174, 'μ2': 0.233, 'μ3': 0.188, 'μ4': 0.5,    'μ5':0.5366, 'μ6': 0.1844, 'μ7': 0.5303, 'μ8': 0.19},
    'C5' : {'μ1': 0.4808, 'μ2': 0.1964, 'μ3': 0.1514, 'μ4': 0.4634, 'μ5': 0.5,    'μ6': 0.1478, 'μ7': 0.4937, 'μ8': 0.1534},
    'C6' : {'μ1': 0.833,  'μ2':0.5486, 'μ3': 0.5036,  'μ4': 0.8156, 'μ5': 0.8522, 'μ6': 0.5,    'μ7': 0.8459, 'μ8': 0.5056},
    'C7' : {'μ1': 0.4871, 'μ2': 0.2027, 'μ3': 0.1577, 'μ4': 0.4697, 'μ5': 0.5063, 'μ6': 0.1541, 'μ7': 0.5,    'μ8': 0.1597},
    'C8' : {'μ1': 0.8274, 'μ2': 0.543, 'μ3': 0.498, 'μ4': 0.81, 'μ5': 0.81, 'μ6': 0.498, 'μ7': 0.81, 'μ8': 0.5}
}
C8= {
    'C1': {'μ1': 0.5,    'μ2': 0.3505, 'μ3': 0.2103,  'μ4': 0.6218, 'μ5': 0.2923, 'μ6': 0.2115,  'μ7': 0.4941, 'μ8': 0.2107},
    'C2': {'μ1': 0.6495, 'μ2': 0.5,    'μ3':0.3598, 'μ4': 0.7713, 'μ5': 0.4418, 'μ6':0.361, 'μ7': 0.6436, 'μ8': 0.3602},
    'C3': {'μ1': 0.7897,  'μ2': 0.6402, 'μ3': 0.5,    'μ4': 0.9115, 'μ5':0.5819, 'μ6': 0.5012,  'μ7': 0.7838, 'μ8': 0.5004},
    'C4': {'μ1': 0.3782, 'μ2': 0.2287, 'μ3': 0.0885, 'μ4': 0.5,    'μ5':0.1705, 'μ6': 0.0897, 'μ7': 0.3723, 'μ8': 0.0889},
    'C5' : {'μ1': 0.7077, 'μ2': 0.5582, 'μ3': 0.4181, 'μ4': 0.8295, 'μ5': 0.5,    'μ6': 0.4192, 'μ7': 0.7018, 'μ8': 0.4185},
    'C6' : {'μ1': 0.7885,  'μ2':0.639, 'μ3': 0.4988,  'μ4': 0.9103, 'μ5': 0.5808, 'μ6': 0.5,    'μ7': 0.7826, 'μ8': 0.4993},
    'C7': {'μ1': 0.5059, 'μ2': 0.3564, 'μ3':0.2162, 'μ4': 0.6277, 'μ5': 0.2982, 'μ6': 0.2174, 'μ7': 0.5,    'μ8': 0.2167},
    'C8': {'μ1': 0.7893, 'μ2': 0.6398, 'μ3': 0.4996, 'μ4': 0.9111, 'μ5':0.5815,  'μ6': 0.5007, 'μ7': 0.7833, 'μ8': 0.5}
}

df_C1 = pd.DataFrame(C1).T
df_C2 = pd.DataFrame(C2).T
df_C3= pd.DataFrame(C3).T
df_C4= pd.DataFrame(C4).T
df_C5= pd.DataFrame(C5).T
df_C6= pd.DataFrame(C6).T
df_C7= pd.DataFrame(C7).T
df_C8= pd.DataFrame(C8).T
coverings = [df_C1, df_C2, df_C3, df_C4, df_C5, df_C6, df_C7, df_C8]

# Fuzzy set A over U
A = {
    'μ1': 0.50,
    'μ2': 0.60,
    'μ3': 0.70,
    'μ4': 0.44,
    'μ5': 0.35,
    'μ6': 0.58,
    'μ7': 0.65,
    'μ8': 0.30
}

# β-co-FN for a single μ and a covering
def beta_co_FN(mu, beta, df_C):
    selected = df_C[df_C[mu] < beta]
    if selected.empty:
        return {e: 1.0 for e in df_C.columns}
    union = selected.max(axis=0)
    return (1 - union).to_dict()

# Aggregate lower approximation over all coverings
def aggregate_lower(A, beta, coverings):
    result = {}
    for mu in A:
        if A[mu] < beta:
            result[mu] = 0.0
        else:
            inner_vals = []
            for df_C in coverings:
                co_FN = beta_co_FN(mu, beta, df_C)
                val = min([(1 - co_FN[nu]) if (1 - co_FN[nu]) > A[nu] else A[nu] for nu in A])
                inner_vals.append(val)
            result[mu] = round(min(inner_vals), 2)
    return result

# Aggregate upper approximation over all coverings
def aggregate_upper(A, beta, coverings):
    result = {}
    for mu in A:
        if A[mu] > 1 - beta:
            result[mu] = 1.0
        else:
            inner_vals = []
            for df_C in coverings:
                co_FN = beta_co_FN(mu, beta, df_C)
                val = max([co_FN[nu] if co_FN[nu] < A[nu] else A[nu] for nu in A])
                inner_vals.append(val)
            result[mu] = round(max(inner_vals), 3)
    return result

# Print fuzzy set output
def print_fuzzy_approximation(name, fuzzy_set):
    print(f"{name} =")
    print(" + ".join([f"{val}/ {mu}" for mu, val in fuzzy_set.items()]) + "\n")
print("β-co-FN eM^β_C^i(μ) for all μ ∈ U (β = 0.5):\n")
for i, df_C in enumerate(coverings, start=1):
    print(f"Covering C^{i}:")
    for mu in U:
        result = beta_co_FN(mu, beta, df_C)
        print(f"  μ = {mu}:")
        for e in U:
            print(f"    {e}: {result[e]:.2f}")
        print()

# Compute results
lower_agg = aggregate_lower(A, beta, coverings)
upper_agg = aggregate_upper(A, beta, coverings)

# Output
print_fuzzy_approximation("Aggregate Lower Approximation Apr_{∑M_C^i}(A)", lower_agg)
print_fuzzy_approximation("Aggregate Upper Approximation Apr_{∑M_C^i}(A)", upper_agg)
```
